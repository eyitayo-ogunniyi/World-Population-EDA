## **Exploratory Data Analysis**

Exploratory Data Analysis(EDA) is an integral part of working with data. It can help to identify errors, as well as better understand patterns within the data, detect outliers or anomalous events, find interesting relations among the variables.

In this project, I performed EDA on world population data. The main purpose is to see how population has changed in each continent. This dataset was downloaded from [kaggle](#https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset).

```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the world population data

world_population = pd.read_csv(r"C:\Users\DELL\Downloads\world_population.csv")
world_population
```

```python
#formatting data

pd.set_option('display.float_format', lambda x: '%.2f' % x)
world_population
```

The dataset contains **234 rows** and **17 columns**. 

We will display the information about the dataset using the *info()* method

```python
#checking the summary of the dataset

world_population.info()
```

```python
#checking all columns that are strings

world_population.select_dtypes(include='object')
```

```python
#checking all columns that are numbers

world_population.select_dtypes(include='number')
```

```python
# checking for null values, non-null values, % of null values, unique count breakdown

def world_population_info():
    pop = pd.DataFrame(index=world_population.columns)
    pop['Datatype'] = world_population.dtypes
    pop['Non Null Values'] = world_population.count()
    pop['Null Values'] = world_population.isna().sum()
    pop['Percentage of Null Values'] = (world_population.isna().mean()) * 100
    pop['Unique Count'] = world_population.nunique()
    return pop

world_population_info()
```

```python
#checking rows with missing values

world_population.loc[world_population.isna().any(axis=1)]
```

```python
#plotting missing values

world_population.isna().sum().plot(kind='barh')
plt.gca().invert_yaxis()

plt.rcParams['figure.figsize'] = (15,7)
```

From the analysis above: **Area, Density, Growth rate, and 1970 - 2022 population** columns have missing values. 

Now, let's check the overview of the numeric variables using *describe()*

```python
#checking the summary statistics for numerical columns

world_population.describe()
```

```python
# checking if there is correlation between numerical columns
world_population.corr(numeric_only=True)
```

```python
#creating heatmap

sns.heatmap(world_population.corr(numeric_only=True), annot=True)
plt.show()
```

From the heatmap above, the population columns shows a strong positive correlation

```python
#checking for outliers using boxplots

world_population.boxplot()
plt.xticks(rotation = 90)
plt.show()
```

Since we are exploring countries with varying population, outliers can be expected

### **Top 10 populated countries**

```python
#checking the top 10 populous countries in 2022

world_population.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
#plotting the 10 most populous countries in 2022

top_10 = world_population.groupby('Country')['2022 Population'].sum().sort_values(ascending=False).head(10)
top_10.plot(kind="barh")

# Adding title
plt.title('Top 10 countries with the largest population in 2022(in billions)', fontsize=14)
plt.gca().invert_yaxis()
```

The most populous country in 2022 is China, followed by India. This result is consistent with report published by [Statista](#https://www.statista.com/statistics/262879/countries-with-the-largest-population/)

Let's check the least populated countries in 2022 and also see the countries with the largest population in 1970

```python
#checking the 10 least populated countries in 2022

world_population.sort_values(by="2022 Population", ascending=False).tail(10)
```

```python
#checking the top 10 populous countries in 1970

world_population.sort_values(by="1970 Population", ascending=False).head(10)
```

```python
#plotting the 10 most populous countries in 1970

top_country = world_population.groupby('Country')['1970 Population'].sum().sort_values(ascending=False).head(10)
top_country.plot(kind="barh")

# Adding title
plt.title('Top 10 countries with the largest population in 1970(in millions)', fontsize=14)
plt.gca().invert_yaxis()
```

**China, India and the United States** were the largest population in 1970 and they were in 2022 too

Now, let's check the summary of the dataset by the continent

```python
#checking summary by continent

grouped = world_population.groupby('Continent').mean(numeric_only=True)

grouped.sort_values(by='2022 Population', ascending=False)
```

```python
#checking the countries in Oceania

world_population[world_population['Continent'] == 'Oceania']
```

### How has population changed over the past years?

```python
#calculate population by each continent

year = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population' , '2022 Population' ]
continent_pop = world_population.groupby('Continent')[year].mean()

world_pop_sorted = continent_pop.sort_values(by='2022 Population', ascending=False)
world_pop_sorted
```

```python
world_pop_sorted = world_pop_sorted.transpose()
world_pop_sorted
```

```python
#creating a line chart
world_pop_sorted.plot()

# Adding title
plt.title('Population Change by Continent (in millions)', fontsize=14)
```

From the above chart, we see that the population of the continent Asia has significantly increased in the past years compared to other continents
